```python
users = [
    {"id": 1, "nome": "Alice", "saldo": 1500.0, "objetivo": "Viagem"},
    {"id": 2, "nome": "Bruno", "saldo": 500.0, "objetivo": "Reserva de Emergência"},
    {"id": 3, "nome": "Carla", "saldo": 10000.0, "objetivo": "Investimentos"}
]
```

```python
def mensagem(users):
  nome = users['nome']
  saldo = users['saldo']
  objetivo = users['objetivo']

  if saldo < 1000:
    return f"Olá {nome}, vimos que você quer montar sua {objetivo}. Que tal começar com poupanças automáticas?"
  elif 1000 < saldo <= 5000:
    return f"Olá {nome}, seu saldo está bom, mas para atingir seu objetivo de {objetivo}, ainda precisa melhorar um pouco, que tal um investimento maior por mês?"
  else:
    return f"Olá {nome}, seu saldo de R$ {saldo:.2f} está excelente! Já pensou em diversificar sua carteira de {objetivo}?"
for user in users:
  user ['mensagem_marketing'] = mensagem(user)
  print(user['mensagem_marketing'])
```

```python
import json

with open('resultado_campanha.json', 'w', encoding='utf-8') as f:
    json.dump(users, f, ensure_ascii=False, indent=4)

print("Processo ETL concluído com sucesso!")
```
